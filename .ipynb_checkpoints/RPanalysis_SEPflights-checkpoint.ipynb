## Importing modules

```python
import numpy as np
import json
import matplotlib
import matplotlib.pyplot as plt
from scipy.stats import kendalltau
from scipy.stats import ttest_ind
```

## Defining functions

```python
def read_armasfile(filename):
    time_moment = []
    altitude = []
    ARMASdose = []
    NAIRASdose = []
    lshell = []
    goes1mev = []
    goes10mev = []
    goes100mev = []
    goeselectr = []
    goessxr = []
    linecounter = 0
    fin = open(filename, 'r')
    for line in fin:
        linecounter += 1
        if (linecounter == 1): continue
        sline = line.split(',')
        time_moment.append(sline[0])
        altitude.append(sline[3])
        ARMASdose.append(sline[4])
        NAIRASdose.append(sline[5])
        lshell.append(sline[11])
        goes1mev.append(sline[12])
        goes10mev.append(sline[15])
        goes100mev.append(sline[16])
        goeselectr.append(sline[13])
        goesflux = sline[17][:-1]
        if (goesflux != ' null'):
            goessxr.append(goesflux)
        else:
            goessxr.append(0.0)
    return time_moment, altitude, ARMASdose, NAIRASdose, lshell, goes1mev, goes10mev, goes100mev, goeselectr, goessxr
```

```python
def plot_flightsummary(filename):
    # reading ARMAS file
    parameters = read_armasfile(filename)
    Lshell = np.array(parameters[4], dtype=float)
    ARMASdoses = np.array(parameters[2], dtype=float)
    NAIRASdoses = np.array(parameters[3], dtype=float)
    protons10mev = np.array(parameters[6], dtype=float)
    sxrflux = np.array(parameters[9], dtype=float)
    # plotting parameters
    matplotlib.rcParams.update({'font.size':15})
    im, ax = plt.subplots(2,1,figsize=(10,15),dpi=160)
    ax[0].plot(Lshell)
    ax[0].set(xlabel='Time, min', ylabel='L-shell', title='Flight '+filename[36:50])
    ax[1].plot(ARMASdoses, label='ARMAS dose rate')
    ax[1].plot(NAIRASdoses, label='NAIRAS dose rate')
    ax[1].set(xlabel='Time, min', ylabel='Dose rate, uSv/hr')
    ax[1].legend()
    ax[2].plot(protons10mev)
    ax[0].set(xlabel='Time, min', ylabel='> 10MeV proton flux, pfu')
    plt.show()
```
